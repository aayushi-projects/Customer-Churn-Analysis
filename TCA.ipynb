### converted 1 and 0 values of seniorcitizen to yes/no to make it easier to understand

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a crosstab
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100  # % row-wise
ct = ct[['No', 'Yes']]  # optional: ensure order

# Step 2: Plot
fig, ax = plt.subplots(figsize=(4, 4))

# Plot 'No' Churn
ax.bar(ct.index, ct['No'], label='No', color='skyblue')

# Plot 'Yes' Churn stacked on top
ax.bar(ct.index, ct['Yes'], bottom=ct['No'], label='Yes', color='salmon')

# Step 3: Add percentage labels
for i in ct.index:
    ax.text(i, ct.loc[i, 'No']/2, f"{ct.loc[i, 'No']:.1f}%", ha='center', va='center', fontsize=8)
    ax.text(i, ct.loc[i, 'No'] + ct.loc[i, 'Yes']/2, f"{ct.loc[i, 'Yes']:.1f}%", ha='center', va='center', fontsize=8)

# Step 4: Styling
ax.set_xticks([0, 1])
ax.set_xticklabels(['Not Senior', 'Senior'])
ax.set_ylabel('% of Customers')
ax.set_title('Churn by SeniorCitizen (Stacked %)')
ax.legend(title='Churn')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots: 3 rows x 3 columns
plt.figure(figsize=(15, 12))

for i, column in enumerate(cols):
    plt.subplot(3, 3, i + 1)
    sns.countplot(x=column, data=df, hue='Churn')
    plt.title(f'Churn by {column}')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

#The subplot analysis reveals that customers who **lack services** like OnlineSecurity, TechSupport, DeviceProtection, and OnlineBackup are **more likely to churn**. Features such as **StreamingTV and InternetService** also show a **higher churn** among users who opted out or have "No internet service." In contrast, customers with **active service subscriptions** tend to show **greater loyalty**. This suggests that offering bundled or value-added services may **reduce customer churn**.
